```python
import matplotlib.pyplot as plt
import numpy as np

# 模擬數據
x_labels = ['Condition A', 'Condition B', 'Condition C']
acc_values_1 = [0.75, 0.80, 0.70]
acc_values_2 = [0.65, 0.85, 0.60]
acc_values_3 = [0.70, 0.78, 0.65]

# x 軸位置
x = np.arange(len(x_labels))

# 條形圖的寬度
width = 0.25

# 繪製長條圖
fig, ax = plt.subplots()

# 在每個 x 軸位置上，繪製多個長條
ax.bar(x - width, acc_values_1, width, label='Method 1')
ax.bar(x, acc_values_2, width, label='Method 2')
ax.bar(x + width, acc_values_3, width, label='Method 3')

# 添加標籤和標題
ax.set_xlabel('Conditions')
ax.set_ylabel('Accuracy')
ax.set_title('Accuracy for Different Conditions and Methods')
ax.set_xticks(x)
ax.set_xticklabels(x_labels)
ax.legend()

# 顯示圖表
plt.show()
```
